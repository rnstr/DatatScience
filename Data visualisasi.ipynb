# Basic Visualization

## Bar Chart

```python
# create data frame
data_1 = {'Country' :['USA','SGP','JPN', 'IDN'], 'GDP' : [40000, 35000, 50000, 25000]}
df1 = pd.DataFrame(data_1, columns = ['Country', 'GDP'])
df1
```

```python
# menggunakan pandas
df1.plot(x= 'Country', y = 'GDP', kind = 'bar')
df1.plot(x= 'Country', y = 'GDP', kind = 'barh')
```

```python
# menggunakan matplotlib
plt.bar(df1['Country'], df1['GDP'])
plt.xlabel('Country')
plt.ylabel('GDP')
plt.title('GDP per Country')
plt.show()
```

## Line Chart

```python
# create data frame
data_2 = {'Year' :[1930,1940,1920,1950], 'GMV' : [40000, 35000, 50000, 25000]}

df2 = pd.DataFrame(data_2, columns = ['Year', 'GMV'])
df2
```

```python
#menggunakan pandas
df2.plot(x = 'Year', y='GMV', kind = 'line')
```

```python
df2_sorted = df2.sort_values('Year', ascending = True)
df2_sorted
```

```python
df2_sorted.plot(x = 'Year', y='GMV', kind = 'line')
```

datanya harus sorted kalau mau pakai line chart

```python
3 #menggunakan matplotlib
plt.plot(df2_sorted['Year'], df2_sorted['GMV'], color = 'red', marker = 'o')
plt.grid()
plt.title('gmv per year')
plt.xlabel('Year', fontsize = 12)
plt.ylabel('GMV', fontsize = 12)
#plt.axes().set_ylim(bottom = 0)
#plt.axes().set_xlim(left=1910)
plt.show()
```

```python
# menggunakan seaborn
sns.lineplot(x=df2_sorted['Year'], y=df2_sorted['GMV'])
plt.grid()
plt.show()
```

## Pie Chart

```python
# create data frame
data_3 = {'Task' :[300,500,700]}
df3 = pd.DataFrame(data_3, columns = ['Task'], index = ['Task Pending', 'Task Ongoing', 'Task Completed'])
df3
```

```python
df3.plot(y='Task' , kind = 'pie' )
```

```python
#menggunakan matplotlib
plt.pie(df3['Task'], labels = ['Task Pending', 'Task Ongoing', 'Task Completed'], autopct = '%.2f%%', colors= ['brown', 'pink', 'purple'])
plt.title('Task Breakdown')
plt.show()
```
