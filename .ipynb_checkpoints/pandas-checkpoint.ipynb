### Pandas

```python
import pandas as pd
```

```python
# Series
apel = pd.Series([5,2,0,1])
print(apel[1])

# DataFrame
dfApel = pd.DataFrame([3,2,0,2], columns=['apel'], index = [1,2,3,4])
print(dfApel)

#DataFrame 2 kolom
dict1 = { 'apel' : [5,4,3,2,1], 'jeruk' : [1,2,3,4,5]}
df1 = pd.DataFrame(dict1)
print(df1)
```

```python
# menghitung jumlah data
print(df1.count())

#menjumlahkan total data
print(df1.sum())

#statstistik singkat
print(df1.describe())
```

```python
# mengambil subset dari data frame
apel = df1.apel
print(apel)

jeruk_apel = df1[['jeruk', 'apel']]
print(jeruk_apel)
```

```python
# mengambil beberapa baris dari dataFrame
print(df1[0:2])
print(df1[2:])
```

```python
# perbandingan nilai 
print(df1['jeruk'] < 3)
```

```python
# membaca baris pertama
print(df1.head(1))
```

```python
# transapose
print(df1.T)
```
